```python
# Using the employees and projects tables above, answer:
import sqlite3
import pandas as pd
import numpy as np

conn=sqlite3.connect(':memory:')
cursor=conn.cursor()

cursor.execute('''
CREATE TABLE employees(
   id INTEGER PRIMARY KEY,
   name TEXT,
   department TEXT,
   salary REAL,
   age INTEGER,
   city TEXT
   )
''')

employees = [
    (1, 'Alice',   'Data Science',  85000, 28, 'Mumbai'),
    (2, 'Bob',     'Engineering',   92000, 32, 'Bangalore'),
    (3, 'Charlie', 'Data Science',  78000, 25, 'Delhi'),
    (4, 'Diana',   'Marketing',     65000, 30, 'Mumbai'),
    (5, 'Eve',     'Engineering',   95000, 35, 'Bangalore'),
    (6, 'Frank',   'Data Science',  88000, 29, 'Chennai'),
    (7, 'Grace',   'Marketing',     70000, 27, 'Delhi'),
    (8, 'Henry',   'Engineering',   82000, 31, 'Mumbai'),
    (9, 'Iris',    'Data Science',  91000, 33, 'Bangalore'),
    (10,'James',   'Marketing',     60000, 24, 'Chennai')
]

cursor.executemany('INSERT INTO employees VALUES (?,?,?,?,?,?)',employees)

conn.commit()

print("Database Created Successfully")

# Create a second table
cursor.execute('''
CREATE TABLE projects (
    project_id INTEGER PRIMARY KEY,
    project_name TEXT,
    employee_id INTEGER,
    budget REAL
)
''')

projects = [
    (1, 'Customer Churn Model',    1, 50000),
    (2, 'Sales Forecasting',       3, 35000),
    (3, 'Recommendation Engine',   6, 75000),
    (4, 'Fraud Detection',         9, 90000),
    (5, 'Website Redesign',        4, 40000),
    (6, 'Mobile App',              2, 120000)
]

cursor.executemany(
    'INSERT INTO projects VALUES (?,?,?,?)',
    projects
)
conn.commit()

def query(sql):
  return pd.read_sql_query(sql,conn)
# TASK 1:
# Find all Data Science employees earning more than 80000
# Show: name, salary, city
print(query('''SELECT name,salary,city FROM employees
WHERE salary > 80000
AND department = "Data Science"
'''))
# TASK 2:
# Find the average salary per city
# Only show cities where average salary > 75000
# Order by average salary descending
print(query('''
SELECT
avg(salary) as average_salary ,
city
FROM employees
GROUP BY city
HAVING avg(salary) > 75000
ORDER BY average_salary DESC
'''))
# TASK 3:
# Find how many employees are in each department
# Show: department, headcount, total_salary, avg_salary
print(query('''
SELECT
department,
COUNT(*) AS headcount,
SUM(salary) AS total_salary,
avg(salary) as average_salary
FROM employees
GROUP BY department
'''))

# TASK 4:
# Find the highest paid employee in each department
# Show: department, name, salary
print(query('''SELECT
e.department,
e.name,
e.salary
FROM employees e
WHERE salary=(
SELECT MAX(salary)
FROM employees
WHERE department = e.department)
'''))

# TASK 5:
# Using JOIN — find all employees who have projects
# Show: name, department, project_name, budget
# Order by budget descending
print(query('''
    SELECT
    e.name,
    e.department,
    p.project_name,
    p.budget
    FROM employees e
    JOIN projects p
     ON e.id=p.project_id
    ORDER BY P.budget DESC

'''))
# TASK 6:
# Using CASE WHEN — categorise all employees as:
# "High" if salary > 85000
# "Medium" if salary between 70000 and 85000
# "Low" if salary < 70000
# Count how many fall in each category
print(query('''
SELECT
    CASE
        WHEN salary > 85000 THEN 'High'
        WHEN salary BETWEEN 70000 AND 85000 THEN 'Medium'
        ELSE 'Low'
    END AS salary_category,
    COUNT(*) AS employee_count
FROM employees
GROUP BY salary_category;
'''))
```
